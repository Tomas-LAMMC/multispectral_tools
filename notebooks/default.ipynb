```python
from micasense.image import Image
import os

image_path = os.path.join('..', 'data', 'tabakas_ground', 'IMG_0008_1.tif')
img = Image(image_path)


print(f"Library path: {os.path.dirname(micasense.__file__)}")
# Display the actual metadata object contents
print(f"Bands: {img.meta.band_name()}")
print(f"Dimensions: {img.meta.image_size()}")
print(f"Camera Model: {img.meta.camera_model()}")
```

```python
from micasense.capture import Capture

# Create a list of paths for all 6 bands
image_names = [f'IMG_0008_{i}.tif' for i in range(1, 7)]
image_paths = [os.path.join('..', 'data', 'tabakas_ground', name) for name in image_names]

# Attempt to create a capture
try:
    cap = Capture.from_filelist(image_paths)
    print("Capture created successfully!")
    print(f"Number of images in capture: {len(cap.images)}")
except Exception as e:
    print(f"Error creating capture: {e}")
```

```python
import micasense.alignment as alignment

# This will likely fail or give weird results because of the size difference
try:
    print("Attempting to calculate alignment...")
    # We use the Pan band (index 5) as the reference
    warp_matrices, alignment_stats = alignment.align_capture(cap, reference_index=5)
    print("Alignment finished!")
except Exception as e:
    print(f"Alignment Failed: {e}")
```
